```python
# Scrape webpages for knowledge articles
for kb in kb_ids:
    kb = "KB" + str(kb).zfill(8)
    url = base_url + kb

    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'lxml')
    result.append(soup)

    file_path = kb + ".html"
    
    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(str(soup)) 

    time.sleep(2)
```

```python
html = []

for filename in os.listdir('C:\\Users\\Chris\\RAG\\data\\cab\\html'):
    if filename.endswith('.html'):
        with open(os.path.join('C:\\Users\\Chris\\RAG\\data\\cab\\html', filename)) as f:
            soup = BeautifulSoup(f, 'html.parser')
            html.append([filename, soup])
```

```python
all_dict = {}
id = 0

for item in html:
    kb_id, extension = os.path.splitext(item[0])
    html = item[1]

    a_dict = {}
    title = html.select_one('div.article__title h1')

    if title is None:
        next
    else:
        title = html.select_one('div.article__title h1').get_text()
        article = html.select_one('div.article__content').get_text()
        categories = html.select_one('div.article__categories')

        a_dict['question'] = title
        a_dict['answer'] = article
        a_dict['kb_id'] = kb_id

        all_dict[id] = a_dict
    id = id + 1

    print(str(id) + " done")
```

```python
# Save to JSON
with open('data.json', 'w') as f:
    json.dump(all_dict, f)
```

```python
# Save to markdown

# Load the JSON file
with open('C:\\Users\\Chris\\RAG\\data\\cab\\data.json', 'r') as file:
    data = json.load(file)

# Iterate through each entry in the JSON file
for entry in data:
    
    question = data[entry]['question']
    answer = data[entry]['answer']
    kb_id = data[entry]['kb_id']

    # Create a markdown string with the attributes
    markdown_content = f"{question}\n\n{answer}"

    # Save the markdown content to a file
    filename = f"{kb_id}.md"  # You can customize the filename as per your requirement

    with open(filename, 'w') as markdown_file:
        markdown_file.write(markdown_content)

print("Markdown files created successfully.")
```
